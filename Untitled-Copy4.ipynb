The general idea behind web scraping is to locate and extract data that exist on a website, and convert them into a format that is usable for analysis. Data contained in webpages are rendered by web browsers (e.g., Firefox, Chrome) from HTML and CSS code. In web scraping, the data that we wish to obtain are often within the HTML code. Thus, our job is to learn how to read the code and effectively extract said data from the HTML.

We will start by reading in the source code of a website and examining it. This can be done by creating a BeautifulSoup object with the BeautifulSoup function.

So, for our first line of code, letâ€™s do that. First we will import the functions necessary for web scraping. Then, we will access the URL of interest and pull its source code. Finally, we will examine its components.

````python
#This code is to import the necessary functions into our Python environment
from bs4 import BeautifulSoup
import urllib

#This accesses the URL of interest and pulls its source code.
r = urllib.urlopen('https://us.sagepub.com/en-us/nam/administrative-science-quarterly/journal202065#editorial-board').read()
soup = BeautifulSoup(r)

#This displays the source code
print soup.prettify()[44900:46000]
````

This is what the source code looks like.  It may look daunting, but it is surprisingly simple to break down.

1) The HTML tags contained in the angled brackets ("<...>" or "</...>") provide structural information (and sometimes formatting) for the website. Usually, we probably don't care about this structural information in and of itself but it is extremely useful for selecting only the content relevant to our needs. <br>
2) Often, the desired content is located in between the beginning of an angled bracket ("<...>") and the end of an angled bracket ("</...>"). For example, the text "Christine Beckman" is located within the angled brackets "< a ...>" and "< /a>" <br>
```
                <a data-contributor="626701" data-sequence="186">
                 Christine Beckman
                </a>
```

3) Notice, these angled brackets are displayed in a hierarchy, where some brackets are nested within others.

This is one way to visualize these hierarchical structures. Each tag is nested within another.  

Ultimately, effective web scraping is about identifying the unique nesting structure behind the data you want to obtain.  Most modern websites make this <i> extremely </i> easy by giving each tag a unique attributes.  

```
Example tags:
    < a> < /a>
    < td> < /td>
    
Example attributes that define each tag:
    < a id = "Link to next page"> < /a>
    < td class = "EditorMember"> < /td>
```

As a result, as long as you know the unique attributes of the data you wish to obtain, you can tell computers to "fetch" all the data contained within any tag containing your unique attribute.

As an example, lets say you want to obtain all the editorial board members for ASQ.  If you look at the above source code, you'll notice that the text containing the editorial board members names are surrounded by tags such as "< td>", and that these tags contain attributes such as "< td class="ed-board-member">". 

```
              <td class="ed-board-member">
                <a data-contributor="721894" data-sequence="184">
                 Caroline Bartel
                </a>
               </td>
               <td class="ed-board-member-affiliation">
                University of Texas at Austin, USA
               </td>
              </tr>
              <tr>
               <td class="ed-board-member">
                <a data-contributor="626701" data-sequence="186">
                 Christine Beckman
                </a>
               </td>
               <td class="ed-board-member-affiliation">
                University of Maryland,  USA
```

You can identify this by <br>
1) Going into the source code <br>
2) Locate the data you want (Ctrl+F works fine) <br>
3) Look at the brackets that surround your data, and find the pattern of attributes that uniquely define these  brackets.   
 
Once you've identified the unique attributes, most programming packages contain pre-defined functions that allow you to pull the information that are identified by each unique attribute. 

Let's put this work. 

The code below accesses the URL of interest and pulls its source code.  
Then, the function findAll extracts all the HTML related to a given tag and attribute.  

````python
# This accesses the URL of interest and pulls its source code.
r = urllib.urlopen('https://us.sagepub.com/en-us/nam/administrative-science-quarterly/journal202065#editorial-board').read()
soup = BeautifulSoup(r)

#Recall the unique identifier for each board member was "<td class="ed-board-member">"
#So, lets go into our source code (contained in object "soup")... 
#... and obtain all tags labeled "td" with the attribute "class = "ed-board-member""
Board = soup.findAll('td',attrs={'class':"ed-board-member"})
print Board[0:10]
````

Great, we've just selected the HTML that contains all the board-members.  That's a great first step.

Next, let's extract the text out of this HTML. 

````python
#Get text from object 'Board', which contains all editorial board members
for member in Board[0:10]:
    print member.text
````

From here, we would typically place the text into an excel file for subsequent use.

In general, that is the basics of web scraping!  <br>
1) Read HTML into the computer <br>
2) Identify the tags and attributes that uniquely identify your data <br>
3) Tell the computer (with a one or two lines of syntax) to obtain the informatino contained within those tags

# 3) Use the scraper from [1] to collect data for every URL that we collect from [2]. 

Great!  We've got our list of URLs.  Now, let's put the pieces together.

In particular, lets take the scraper we used to collect all the URLs for a given journal issue , and collect all URLs for a given issue [2].  Then, lets visit each URL obtained for that issue (obtained from [2]), and, using the scraper we built in [1], collect the abtract and title.

For this, we'll need to have a closer look at what a for loop does. For loops are an easy way to tell a computer to perform some actions a preset number of times.  For example...

```
list = ['hello','world','1','5','reviewer 2 always hates my paper']

for item in list:
    print item

hello
world
1
5
reviewer 2 always hates my paper
```

The above code will go through every object in our list one at a time...  <br>
During the first iteration, it will take the object 'hello' and assign it to object 'item'.  It then prints object 'item'.  <br>
During the second iteration, it will take the object 'world' and assign it to object 'item'.  It then prints object 'item'. <br>
During the third iteration, it will take object '1' and assign it to object 'item'.  It then prints object 'item'. <br>
...This process continues until the for loop reaches the end of the list.

----

In this case, we will use this for loop to visit each webpage.  

Recall that all our links are stored in the list 'cleanLinkSet'.  We can tell the computer the following:

```
for every link in the list 'cleanLinkSet'...
    Construct the URL for the webpage we want to visit
    Use the scraper we designed in [1]
    
```

This tells the computer to visit each webpage in the list 'cleanLinkSet', and pull the data we want from said webpage.  

Why don't we try that below? 
The below code does the following:

(S) Sets up the key variables we want (a dataframe to store our data, and the base URL) <br>
(U) Use the scraper we designed in [2] to obtain the links to all articles from a given  journal issue <br>
(A) For every link... use the scraper we designed in [1] and obtain the abstract and title

````python
############
# (S) Setup 
############

### (S0) Initialize key variables

#Create new DataFrame
data = pandas.DataFrame([])

#Set Base URL
baseURL = 'http://amj.aom.org'

############
# (U) Obtain URLs for each article in a given issue
#
# This is the scraper we designed in [2] to visit the webpage of a 
# given issue and pull all URLs for abstracts
############

### (U0) Access a given URL for an issue; obtain HTML containing links to abstracts

#Access the URL for a given issue and obtain its source code
r = urllib.urlopen('http://amj.aom.org/content/59/5.toc').read()
soup = BeautifulSoup(r)

#Go into object "soup" (the source code) and then find all URLs to abstracts
#... recall these URLs were contained in objects 'a' with attributes "rel" = "abstract" 
rawLinkSet = soup.findAll('a',attrs={'rel':"abstract"})

### (U1) Take the HTML containing links to abstracts, and place links into a list

#Create empty list
cleanLinkSet = []

#For every link in list "rawLinkSet"
for link in rawLinkSet:
    
    #Take the 'href' from that link and append it to our list
    cleanLinkSet.append(link['href'])

############
# (A) Obtain abstracts and title from each URL
#
# This is the scraper we designed in [1] to collect the
# abstract and title from a given article 
############

# (A0) Create loop to visit each article in a given issue and obtain its data

#For every link in my set of links
for link in cleanLinkSet:
    
    #Construct URL for article
    URL = baseURL+link
   
    #Read HTML from that URL
    r = urllib.urlopen(URL).read()
    soup = BeautifulSoup(r)
    
    #Obtain abstract
    #...recall abstracts were contained in object "div" with attributes "id" = "abstract-1"
    abstract = soup.find('div',attrs={'id':"abstract-1"})
    abstract = abstract.getText()
    
    #Clean text
    abstract = abstract.replace("\nAbstract\n","")
    abstract = abstract.replace("\n","")
    abstract = abstract.replace("  ","")
    abstract = abstract.encode('utf-8').decode('ascii', 'ignore')
    
    #Obtain Title
    #...recall titles were contained in object "h1" with attributes "id" = "article-title-1"
    title = soup.find('h1',attrs={'id':"article-title-1"})
    title = title.getText()
    
    #Populate DataFrame
    data = data.append([[abstract,title]])
    
print data
````

Great!  We've just obtained all the abstracts and titles for one issue of AMJ.
Abstracts are located in column 0, and all titles in column 1.

Now, if we assume that all webpages have the same HTML layout (and, in particular, the information we want is surrounded by the same tag with the same unique attributes), then we can do the following:

```
Take the link to a journal issue we want to start with

While we are not on the last issue...

    Take the HTML from a particular journal issue     
    Use the scraper we designed in [2] to obtain the links for each article in the issue
    
        for every link obtained from [2]...        
            Use the scraper we designed in [1] to obtain the abstract and title
            Store the data
    
    Check if we are on the last issue...    
        If yes, kill the while loop
        
    Otherwise, go obtain the link associated with the previous issue 
```

````python
import time

#Pause your script for 15 seconds
time.sleep(15)
````

Okay, with this in mind, lets give this script a shot!

The script below does the following

* (S) Sets up the key variables we want (a dataframe to store our data, and the base URL) <br>
* (C) Crawl through a given journal issue until we hit the last issue
    * While we are not on the last journal issue...
    * (U) Use the scraper we designed in [2] to obtain the links to all articles from a given  journal issue <br>
    * (A) For every link... use the scraper we designed in [1] and obtain the abstract and title
    * (C) Check if we're on the last issue.  If we are, stop the loop.  If we are not, get the link associated with 'previous issue'

````python
############
# (S) Setup 
############

### (S0) Initialize key variables

#Create new DataFrame
data = pandas.DataFrame([])

#Set Base URL
baseURL = 'http://amj.aom.org'

#URL for the starting issue
IssueURL = 'http://amj.aom.org/content/59/5.toc'

############
# (C) Crawl 
#
# This is the scraper we designed in [2] to visit the webpage of a 
# given issue and pull all URLs for abstracts
############

# (C0) While we are NOT on the first issue

while True:
    
    ############
    # (U) Obtain URLs for each article in a given issue
    #
    # This is the scraper we designed in [2] to visit the webpage of a 
    # given issue and pull all URLs for abstracts
    ############

    ### (U0) Access a given URL for an issue; obtain HTML containing links to abstracts

    #Access the URL for a given issue and obtain its source code
    r = urllib.urlopen(IssueURL).read()
    soup = BeautifulSoup(r)

    #Go into object "soup" (the source code) and then find all URLs to abstracts
    #... recall these URLs were contained in objects 'a' with attributes "rel" = "abstract" 
    rawLinkSet = soup.findAll('a',attrs={'rel':"abstract"})

    ### (U1) Take the HTML containing links to abstracts, and place links into a list

    #Create empty list
    cleanLinkSet = []

    #For every link in list "rawLinkSet"
    for link in rawLinkSet:

        #Take the 'href' from that link and append it to our list
        cleanLinkSet.append(link['href'])

    ############
    # (A) Obtain abstracts and title from each URL
    #
    # This is the scraper we designed in [1] to collect the
    # abstract and title from a given article 
    ############

    # (A0) Create loop to visit each article in a given issue and obtain its data

    #For every link in my set of links
    for i in range(0,len(cleanLinkSet)):

        #Construct URL for article
        URL = baseURL+cleanLinkSet[i]

        #Read HTML from that URL
        r = urllib.urlopen(URL).read()
        soup = BeautifulSoup(r)

        #Obtain abstract
        #...recall abstracts were contained in object "div" with attributes "id" = "abstract-1"
        abstract = soup.find('div',attrs={'id':"abstract-1"})
        abstract = abstract.getText()

        #Clean text
        abstract = abstract.replace("\nAbstract\n","")
        abstract = abstract.replace("\n","")
        abstract = abstract.replace("  ","")
        abstract = abstract.encode('utf-8').decode('ascii', 'ignore')

        #Obtain Title
        #...recall titles were contained in object "h1" with attributes "id" = "article-title-1"
        title = soup.find('h1',attrs={'id':"article-title-1"})
        title = title.getText()

        #Populate DataFrame
        data = data.append([[abstract,title]])
        
    # (C1) Check if we are on the last URL
    
    #If the IssueURL is equal to the URL of the last issue
    if IssueURL == 'http://amj.aom.org/content/1/1.toc':
        
        #Kill the loop.
        break
    
    #Otherwise get the URL for the previous issue.
    else:
        
        #Visit IssueURL and obtain source code
        r = urllib.urlopen(IssueURL).read()
        soup = BeautifulSoup(r)
        
        #Get the HTML associated with the URL for the prior issue
        IssueURL = soup.find('a',attrs={'title':"Previous issue"})
        
        #Extract href from HTML
        IssueURL = IssueURL['href']
        
        #Construct URL
        IssueURL = baseURL + IssueURL
        
    #Wait for 15 seconds so you do not slam AMJ's servers
    time.sleep(15)

        
    
#Once the loop exits, print the data so we can look at it.    
print data
````

````python
def getAbstractTitle(data,baseURL,URL):
    '''
    Input: a pandas DataFrame containing two columns, a base URL, and a URL
    Output: pandas Dataframe containing two unlabeled columns: [Abstract, Title]
    
    Algorithm:
    Construct URL for article
    Read HTML from URL
    Obtain and clean abstract
    Obtain title
    Return dataframe
    '''
    #Construct URL for article
    URL = baseURL+cleanLinkSet[i]

    #Read HTML from that URL
    r = urllib.urlopen(URL).read()
    soup = BeautifulSoup(r)

    #Obtain abstract
    #...recall abstracts were contained in object "div" with attributes "id" = "abstract-1"
    abstract = soup.find('div',attrs={'id':"abstract-1"})
    abstract = abstract.getText()

    #Clean text
    abstract = abstract.replace("\nAbstract\n","")
    abstract = abstract.replace("\n","")
    abstract = abstract.replace("  ","")
    abstract = abstract.encode('utf-8').decode('ascii', 'ignore')

    #Obtain Title
    #...recall titles were contained in object "h1" with attributes "id" = "article-title-1"
    title = soup.find('h1',attrs={'id':"article-title-1"})
    title = title.getText()

    #Populate DataFrame
    data = data.append([[abstract,title]])
    
    #Return
    return(data)

def getAbstractLinks(IssueURL):
    '''
    Input: URL for an issue
    Output: A list of URL links for each article in the issue
    
    Algorithm:
    Access IssueURL and pull source code
    Obtain HTML containing URL for each abstract
    Create empty list
    Populate list by pulling the href associated with each URL
    Return list
    '''
    #Access the URL for a given issue and obtain its source code
    r = urllib.urlopen(IssueURL).read()
    soup = BeautifulSoup(r)

    #Go into object "soup" (the source code) and then find all URLs to abstracts
    #... recall these URLs were contained in objects 'a' with attributes "rel" = "abstract" 
    rawLinkSet = soup.findAll('a',attrs={'rel':"abstract"})

    #Create empty list
    cleanLinkSet = []

    #For every link in list "rawLinkSet"
    for link in rawLinkSet:

        #Take the 'href' from that link and append it to our list
        cleanLinkSet.append(link['href'])
        
    return(cleanLinkSet)

def getPriorIssueURL(IssueURL):
    '''
    Input: URL for current issue
    Output: URL for previous issue
    
    Algorithm:
    Access the URL for a given issue and obtain its source code
    Get HTML associated with URL for prior issue
    Obtain href from HTML
    Return URL    
    '''
    #Access the URL for a given issue and obtain its source code
    r = urllib.urlopen(IssueURL).read()
    soup = BeautifulSoup(r)
    
    #Get the HTML associated with the URL for the prior issue
    IssueURL = soup.find('a',attrs={'title':"Previous issue"})
        
    #Extract href from HTML
    IssueURL = IssueURL['href']
    
    return(IssueURL)


#############################################
#        Final script to scrape data        #
#############################################

############
# (S) Setup 
############

### (S0) Initialize key variables

#Create new DataFrame
data = pandas.DataFrame([])

#Set Base URL
baseURL = 'http://amj.aom.org'

#URL for the starting issue
IssueURL = 'http://amj.aom.org/content/59/5.toc'

############
# (C) Crawl 
#
# This is the scraper we designed in [2] to visit the webpage of a 
# given issue and pull all URLs for abstracts
############

# (C0) While we are NOT on the first issue

while FirstIssue = False:
    
    ### Execute crawl
    #Obtain abstract URLs from issue
    abstractLinks = getAbstractLinks(IssueURL)
    
    #For all links in list
    for links in abstractLinks:
        #Get abstract and title
        data = getAbstractTitle(data,baseURL,links)
        
    #If the IssueURL is equal to the URL of the last issue
    if IssueURL == 'http://amj.aom.org/content/1/1.toc':        
        #Kill the loop.
        break
    
    #Otherwise get the URL for the previous issue.
    else:
        #Get URL from prior issue
        IssueURL = getPriorIssueURL(IssueURL)
        
    #Wait for 15 seconds so you don't slam AMJ's servers
    time.sleep(15)
````
